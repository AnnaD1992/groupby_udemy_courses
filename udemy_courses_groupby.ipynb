```python
(data
.groupby("level")
[["is_paid", "price", "num_subscribers"]]
.mean()
)
```

```python
(data
.groupby("level")
[["is_paid", "price", "num_subscribers"]]
.count()
)
```

```python
(data
.groupby("level")
[[ "num_reviews"]]
.mean()
.reset_index()
)
```

```python
grouped=data.groupby("level")["price"]
grouped.agg([np.mean, np.sum])
```

```python
(data.groupby("level")["price"].agg([np.mean, np.sum, np.std ]))
```

```python
data.groupby("level").agg(
        min_price=("price", "min"),
        max_price=("price", "max"),
        average_num_subscribers=("num_subscribers", np.mean)
)
```

```python
data.groupby("level")["num_reviews"].agg(
        min_reviews="min",
        max_reviews="max"
)
```

```python
data.groupby("level").agg(
    {"price":sum,
    "num_subscribers": np.mean}
)
```

```python
data.groupby("level").agg(
    {"price":"sum",
    "num_subscribers": "mean",
    "num_reviews":"std"}
)
```

```python
data.groupby("level").agg(
    {
    "num_subscribers": np.mean}
).plot()
```

```python
data.groupby("level").agg(
    {
    "num_subscribers": np.mean}
).boxplot();
```

```python
data.groupby("level").agg(
    {
    "num_subscribers": "mean",
    "num_reviews":"std"}
).boxplot()
```
